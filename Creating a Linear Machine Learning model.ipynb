```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
obs = 1000

xs = np.random.uniform(-10, 10, (obs,1))
zs = np.random.uniform(-10, 10, (obs,1))

inputs = np.column_stack((xs,zs))
```

```python
#Add Noise to the function to make the function random while preserving the linearity of the model

noise = np.random.uniform(-1, 1, (obs,1))

targets = 2*xs - 3*zs + 5 + noise
```

### Plotting the training data (Not necessary, just for visualization)

```python
targets = targets.reshape(obs,)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_xlabel('sz')
ax.set_xlabel('targets')
ax.view_init(azim=200)
plt.show()

targets = targets.reshape(obs,1)
```

### Setting initial values

```python
init_value  = 0.1

weight = np.random.uniform(-init_value, init_value, size=(2,1))
bias = np.random.uniform(-init_value,init_value, size=1)
```

### Selecting Optimal Learning Rate

```python
learning_rate = 0.02
```

### Training the model

```python
#1. Create a for loop to iterate through 100 instances
#2. At each iteration we calculate the output
#3. Compare outputs to target through the loss
#4. output the loss
#5. Adjust weights and bias to know how well the model is performing.

for i in range(100):
    outputs = np.dot(inputs, weight) + bias
    delta = outputs - targets
    loss = np.sum(delta**2)/ 2 / obs
    
    print(loss)
    
    scaled_delta = delta / obs
    
    weight = weight - learning_rate*np.sum(np.dot(inputs.T, scaled_delta))
    bias = bias - learning_rate*np.sum(scaled_delta)
```
