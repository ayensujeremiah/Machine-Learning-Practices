```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
obs = 1000

xs = np.random.uniform(-10, 10, (obs,1))
zs = np.random.uniform(-10, 10, (obs,1))

inputs_generated = np.column_stack((xs,zs))

noise = np.random.uniform(-1, 1, (obs,1))

targets_generated = 2*xs - 3*zs + 5 + noise

np.savez('TF_intro', inputs=inputs_generated, targets=targets_generated)
```

### Loading Tensor

```python
training_data = np.load('TF_intro.npz')

input_size = 2
output_size = 1

ki = tf.random_uniform_initializer(minval=-0.1, maxval=0.1)
bi =  tf.random_uniform_initializer(minval=-0.1, maxval=0.1)
model = tf.keras.Sequential([
                                tf.keras.layers.Dense(output_size, 
                                kernel_initializer=ki,
                                 bias_initializer=bi
                                  )
                           ])

customize_optimizer = tf.keras.optimizers.SGD(learning_rate=0.02)


model.compile(optimizer=customize_optimizer, loss='mean_squared_error')

model.fit(training_data['inputs'], training_data['targets'], epochs=100, verbose=0)
```
